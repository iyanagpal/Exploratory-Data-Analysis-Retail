```python
#Visualizing relation among features depending on Category

sns.pairplot(s_data, hue = 'Category')
plt.show()
```

```python
pd.DataFrame(s_data.groupby(['Ship Mode', 'Segment', 'Category', 'Sub-Category', 'State', 'Region'])['Quantity', 'Discount', 'Sales', 'Profit'].sum().reset_index())
```

```python
#Grouping and aggregation based on States

s_data.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

```python
#Sales_Profit grouping based on State

Sales_Profit = s_data.groupby("State")["Profit"].sum()
Sales_Profit
```

```python
#Graph between State and Profit

plt.figure(figsize = (20, 10))
sns.barplot(x=Sales_Profit.index, y=Sales_Profit.values)
plt.ylabel("According to State Profit")
plt.xticks(rotation=90)
plt.show()
```

```python
#Sales grouping based on State

Sales = s_data.groupby("State")["Sales"].sum()
Sales.plot.bar(figsize = (15, 10))
plt.show()
```

```python
#Category grouping based on sales and profit

Category = s_data.groupby("Category")["Profit","Sales"].sum()
Category.plot.bar(figsize = (15, 10))
plt.show()
```

```python
#Sub-Category grouping based on sales and profit

Sub_Category = s_data.groupby("Sub-Category")["Profit","Sales"].sum()
Sub_Category.plot.bar(figsize = (15, 10))
plt.show()
```

```python
#State grouping based on sales and profit

State = s_data.groupby("State")["Profit","Sales"].sum()
State.plot.bar(figsize = (15, 10))
plt.show()
```

```python
#Graph of Category based on Region

plt.figure(figsize = (20, 10))
sns.countplot(x= 'Category', hue = 'Region', data = s_data)
plt.xticks(rotation=90)
plt.show()
```

```python
#Graph of Sub-Category based on Region

plt.figure(figsize = (20, 10))
sns.countplot(x= 'Sub-Category', hue = 'Region', data = s_data)
plt.xticks(rotation=90)
plt.show()
```

### 4. Conclusion

1. Profits and sales are highly correlated.
2. There is no correlation among quantity and profit.
3. In Category, Furniture has large number of sales, instead of making profit, it makes loss. So, there is need to limit the sales of Furniture.
4. In Sub-Category, Tables has has large number of sales, instead of making profit, it makes loss. So, there is need to limit the sales of Tables.
5. States like Texas, Ohio has huge sales but it makes big loss. So, there is need to reduce the sales in these states.
6. States like Michigan, Indiana, Virginia, Georgia makes good profit. So, sales can be increased in these states
